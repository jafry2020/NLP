# Preprocessing

```python
processed_data = []

for index, row in df.iterrows():
    
    ghazal = row['Content'].split('|')
    
    for sentence in ghazal:
        new = []
        for word in sentence.split(' '):
            if word != '':
                new.append(word)
        
        processed_data.append(new)

        
processed_data
```

## Generating Vocabulary

```python
list_words = []

for sentence in processed_data:
    if not sentence:
        # print(sentence)
        continue
        
    for word in sentence:
        list_words.append(word)
        
vocabulary = set(list_words)
vocabulary
```

## Generating Unigrams

```python
unigram_freq = {}

for sentence in processed_data:
    if not sentence:
        continue
        
    for word in sentence:
        if word not in unigram_freq:
                unigram_freq[word] = 1

        else:
            unigram_freq[word] = unigram_freq[word] + 1
        
print(len(unigram_freq))

unigram_freq
       
```

## Generating Bigrams

```python
bigram_freq = {}

for sentence in processed_data:
    if len(sentence) < 2 or not sentence:  # Skip sentences with fewer than two words
        continue
        
    for i in range(len(sentence) - 1):
        bigram = (sentence[i], sentence[i + 1])
        if bigram not in bigram_freq:
            bigram_freq[bigram] = 1
        else:
            bigram_freq[bigram] += 1

print(len(bigram_freq))
bigram_freq
```

## Reverse Bigram Freq

```python
reverse_bigram_freq = {}

for sentence in processed_data:
    if len(sentence) < 2 or not sentence:  
        continue
        
    for i in range(len(sentence) - 1):
        
        reverse_bigram = (sentence[i + 1], sentence[i])
        if reverse_bigram not in reverse_bigram_freq:
            reverse_bigram_freq[reverse_bigram] = 1
        else:
            reverse_bigram_freq[reverse_bigram] += 1

print(len(reverse_bigram_freq))
reverse_bigram_freq
```

## Generating Trigrams

```python
trigram_freq = {}

for sentence in processed_data:
    if len(sentence) < 3 or not sentence:  # Skip sentences with fewer than three words
        continue
        
    for i in range(len(sentence) - 2):
        trigram = (sentence[i], sentence[i + 1], sentence[i + 2])
        if trigram not in trigram_freq:
            trigram_freq[trigram] = 1
        else:
            trigram_freq[trigram] += 1

print(len(trigram_freq))
trigram_freq
```
